```python
import tkinter as tk
import random as rd
import numpy as np
class TilesGui:
    def __init__(self, root):
        root.geometry("1900x1900")
        # Create 794 x 1247 pixel
        self.canvas = tk.Canvas(root, width = 1700, height = 1700, background = 'yellow')
        self.plane = self.create_plane()
        self.canvas.pack()    
        self.canvas.bind('<ButtonPress-1>', self.onStart)      
        self.canvas.bind('<B1-Motion>', self.onMove)       
        self.canvas.bind('<ButtonRelease-1>', self.onBreak)
        square_1 = self.cut_square(axis = 1)
        square_2 = self.cut_square(axis = 2)
        square_3 = self.cut_square(axis = 3)
        square_4 = self.cut_square(axis = 4)
        hexagon_1 = self.cut_hexagon(axis = 1)
        hexagon_2 = self.cut_hexagon(axis = 2)
        hexagon_3 = self.cut_hexagon(axis = 3)
        hexagon_4 = self.cut_hexagon(axis = 4)
        squares = [self.create_square(), square_1[0], square_1[1], square_2[0], square_2[1], square_3[0], square_3[1], square_4[0], square_4[1]]
        hexagons = [self.create_hexagon(), hexagon_1[0], hexagon_1[1], hexagon_2[0], hexagon_2[1], hexagon_3[0], hexagon_3[1], hexagon_4[0], hexagon_4[1]]
        self.kinds = [squares, hexagons]
        self.put = []
        self.moved = None
        self.give_coords = []
        ra = self.calculate_remaining_area()
        self.text_ra = self.canvas.create_text(900, 300, text='Remaining area: ' + str(ra[0]))
        self.text_ra_per = self.canvas.create_text(929, 320, text='Remaining area percentage: ' + str(ra[1]))
        self.random_button = tk.Button(root, text="Random Placement", command = lambda threshold = 100000: self.random_put(threshold))
        self.stochastic_button = tk.Button(root, text="Stochastic Placement", command = lambda threshold = 100000: self.stochastic_put(threshold))
        self.random_button.place(x=840, y=350)
        self.stochastic_button.place(x=980, y=350)
        self.rl_button.place(x=1120, y=350)
    def cut_square(self, axis = 1):
        side_1, side_2 = None, None
        if axis == 1:
            side_1 = self.canvas.create_rectangle(1000, 100, 1038, 119, fill='green')
            side_2 = self.canvas.create_rectangle(1000, 129, 1038, 148, fill='green')
        elif axis == 2:
            side_1 = self.canvas.create_rectangle(1100, 100, 1119, 138, fill='orange')
            side_2 = self.canvas.create_rectangle(1139, 100, 1158, 138, fill='orange')
        elif axis == 3:
            side_1 = self.canvas.create_polygon([1200, 100, 1238, 138, 1200, 138], fill='maroon')
            side_2 = self.canvas.create_polygon([1239, 100, 1277, 100, 1277, 138], fill='maroon')
        elif axis == 4:
            side_1 = self.canvas.create_polygon([1300, 100, 1300, 138, 1338, 100], fill='chocolate1')
            side_2 = self.canvas.create_polygon([1377, 100, 1377, 138, 1339, 138], fill='chocolate1')
        return (side_1, "square_cut"), (side_2, "square_cut")
    def cut_hexagon(self, axis = 1):
        side_1, side_2 = None, None
        if axis == 1:
            side_1 = self.canvas.create_polygon([981, 233, 1000, 200, 1038, 200, 1057, 233], fill='green')
            side_2 = self.canvas.create_polygon([1062, 234, 1081, 267, 1119, 267, 1138, 234], fill='green')
        elif axis == 2:
            side_1 = self.canvas.create_polygon([1162, 200, 1181, 200, 1181, 266, 1162, 266, 1143, 233], fill='orange')
            side_2 = self.canvas.create_polygon([1186, 200, 1205, 200, 1224, 233, 1205, 266, 1186, 266], fill='orange')
        elif axis == 3:
            side_1 = self.canvas.create_polygon([1248, 200, 1229, 233, 1248, 266, 1286, 266], fill='maroon')
            side_2 = self.canvas.create_polygon([1291, 200, 1329, 200, 1348, 233, 1329, 266], fill='maroon')
        elif axis == 4:
            side_1 = self.canvas.create_polygon([1372, 200, 1410, 200, 1372, 266, 1353, 233], fill='chocolate1')
            side_2 = self.canvas.create_polygon([1453, 200, 1472, 233, 1453, 266, 1415, 266], fill='chocolate1')
        return (side_1, "hexagon_cut"), (side_2, "hexagon_cut")
    def create_plane(self):
        self.canvas.create_rectangle(0, 0, 794, 1247, fill='white')
    def create_square(self):
        square = self.canvas.create_rectangle(900, 100, 938, 138, fill='blue')
        return square, "square"
    def create_hexagon(self):
        hexagon = self.canvas.create_polygon([900, 200, 938, 200, 957, 233, 938, 266, 900, 266, 881, 233], fill='red')
        return hexagon, "hexagon"
    def check_overlapping(self, shape):
        x1, x2, y1, y2 = self.canvas.coords(shape)
        return self.canvas.find_overlapping(x1, x2, y1, y2)
    def onStart(self, event):
        if 900 <= event.x and event.x <= 938 and 100 <= event.y and event.y <= 138:
            self.moved = self.kinds[0][0]
            self.give_coords = [0, 0]
        elif 1000 <= event.x and event.x <= 1038 and 100 <= event.y and event.y <= 119:
            self.moved = self.kinds[0][1]
            self.give_coords = [0, 1]
        elif 1000 <= event.x and event.x <= 1038 and 129 <= event.y and event.y <= 148:
            self.moved = self.kinds[0][2]
            self.give_coords = [0, 2]
        elif 1100 <= event.x and event.x <= 1138 and 100 <= event.y and event.y <= 138:
            self.moved = self.kinds[0][3]
            self.give_coords = [0, 3]
        elif 1139 <= event.x and event.x <= 1158 and 100 <= event.y and event.y <= 138:
            self.moved = self.kinds[0][4]
            self.give_coords = [0, 4]
        elif 1200 <= event.x and event.x <= 1238 and 100 <= event.y and event.y <= 138:
            self.moved = self.kinds[0][5]
            self.give_coords = [0, 5]
        elif 1239 <= event.x and event.x <= 1277 and 100 <= event.y and event.y <= 138:
            self.moved = self.kinds[0][6]
            self.give_coords = [0, 6]
        elif 1300 <= event.x and event.x <= 1338 and 100 <= event.y and event.y <= 138:
            self.moved = self.kinds[0][7]
            self.give_coords = [0, 7]
        elif 1339 <= event.x and event.x <= 1377 and 100 <= event.y and event.y <= 138:
            self.moved = self.kinds[0][8]
            self.give_coords = [0, 8]
        elif 881 <= event.x and event.x <= 957 and 200 <= event.y and event.y <= 266:
            self.moved = self.kinds[1][0]
            self.give_coords = [1, 0]
        elif 981 <= event.x and event.x <= 1057 and 200 <= event.y and event.y <= 233:
            self.moved = self.kinds[1][1]
            self.give_coords = [1, 1]
        elif 1062 <= event.x and event.x <= 1119 and 234 <= event.y and event.y <= 267:
            self.moved = self.kinds[1][2]
            self.give_coords = [1, 2]
        elif 1143 <= event.x and event.x <= 1181 and 200 <= event.y and event.y <= 266:
            self.moved = self.kinds[1][3]
            self.give_coords = [1, 3]
        elif 1186 <= event.x and event.x <= 1224 and 200 <= event.y and event.y <= 266:
            self.moved = self.kinds[1][4]
            self.give_coords = [1, 4]
        elif 1229 <= event.x and event.x <= 1286 and 200 <= event.y and event.y <= 266:
            self.moved = self.kinds[1][5]
            self.give_coords = [1, 5]
        elif 1291 <= event.x and event.x <= 1348 and 200 <= event.y and event.y <= 266:
            self.moved = self.kinds[1][6]
            self.give_coords = [1, 6]
        elif 1353 <= event.x and event.x <= 1410 and 200 <= event.y and event.y <= 266:
            self.moved = self.kinds[1][7]
            self.give_coords = [1, 7]
        elif 1415 <= event.x and event.x <= 1472 and 200 <= event.y and event.y <= 266:
            self.moved = self.kinds[1][8]
            self.give_coords = [1, 8]
        self.start = event
    def onMove(self, event):                           
        self.canvas = event.widget
        diffX, diffY = (event.x - self.start.x), (event.y - self.start.y)
        if self.moved != None: self.canvas.move(self.moved[0], diffX, diffY)
        self.start = event
    def onBreak(self, event):
        if self.moved != None:
            coords = self.canvas.coords(self.moved[0])
            max_x = max(coords[i] for i in range(0, len(coords), 2))
            min_x = min(coords[i] for i in range(0, len(coords), 2))
            max_y = max(coords[i] for i in range(1, len(coords), 2))
            min_y = min(coords[i] for i in range(1, len(coords), 2))
            if self.moved[1] == "hexagon" or self.moved[1] == "hexagon_cut":
                collision = self.canvas.find_overlapping(min_x + 19, min_y, max_x - 19, max_y)
            else:
                collision = self.canvas.find_overlapping(min_x, min_y, max_x, max_y)
            if 0 <= min_x and max_x <= 794 and 0 <= min_y and max_y <= 1247 and len(collision) == 2:
                if self.moved[1] == "square" or self.moved[1] == "square_cut":
                    if self.give_coords[1] == 0:
                        shape = self.create_square()
                        self.kinds[0][0] = shape
                    else:
                        shape = self.cut_square(axis = (self.give_coords[1] + 1) // 2)[(self.give_coords[1] + 1) % 2]
                        self.kinds[0][self.give_coords[1]] = shape
                elif self.moved[1] == "hexagon" or self.moved[1] == "hexagon_cut":
                    if self.give_coords[1] == 0:
                        shape = self.create_hexagon()
                        self.kinds[1][0] = shape
                    else:
                        shape = self.cut_hexagon(axis = (self.give_coords[1] + 1) // 2)[(self.give_coords[1] + 1) % 2]
                        self.kinds[1][self.give_coords[1]] = shape
                self.put.append(self.moved)
                ra = self.calculate_remaining_area()
                self.canvas.itemconfig(self.text_ra, text='Remaining area: ' + str(ra[0]))
                self.canvas.itemconfig(self.text_ra_per, text='Remaining area percentage: ' + str(ra[1]))
                self.canvas.update()
                self.moved = None
            else:
                if self.moved[1] == "square":
                    self.canvas.move(self.moved[0], 900 - min_x, 100 - min_y)
                elif self.moved[1] == "square_cut":
                    if (self.give_coords[1] + 1) // 2 == 1:
                        if self.give_coords[1] % 2 == 0:
                            self.canvas.move(self.moved[0], 1000 - min_x, 129 - min_y)
                        elif self.give_coords[1] % 2 == 1:
                            self.canvas.move(self.moved[0], 1000 - min_x, 100 - min_y)
                    elif (self.give_coords[1] + 1) // 2 == 2:
                        if self.give_coords[1] % 2 == 0:
                            self.canvas.move(self.moved[0], 1139 - min_x, 100 - min_y)
                        elif self.give_coords[1] % 2 == 1:
                            self.canvas.move(self.moved[0], 1100 - min_x, 100 - min_y)
                    elif (self.give_coords[1] + 1) // 2 == 3:
                        if self.give_coords[1] % 2 == 0:
                            self.canvas.move(self.moved[0], 1239 - min_x, 100 - min_y)
                        elif self.give_coords[1] % 2 == 1:
                            self.canvas.move(self.moved[0], 1200 - min_x, 100 - min_y)
                    elif (self.give_coords[1] + 1) // 2 == 4:
                        if self.give_coords[1] % 2 == 0:
                            self.canvas.move(self.moved[0], 1339 - min_x, 100 - min_y)
                        elif self.give_coords[1] % 2 == 1:
                            self.canvas.move(self.moved[0], 1300 - min_x, 100 - min_y)
                elif self.moved[1] == "hexagon":
                    self.canvas.move(self.moved[0], 881 - min_x, 200 - min_y)
                elif self.moved[1] == "hexagon_cut":
                    if (self.give_coords[1] + 1) // 2 == 1:
                        if self.give_coords[1] % 2 == 0:
                            self.canvas.move(self.moved[0], 1062 - min_x, 234 - min_y)
                        elif self.give_coords[1] % 2 == 1:
                            self.canvas.move(self.moved[0], 981 - min_x, 200 - min_y)
                    elif (self.give_coords[1] + 1) // 2 == 2:
                        if self.give_coords[1] % 2 == 0:
                            self.canvas.move(self.moved[0], 1186 - min_x, 200 - min_y)
                        elif self.give_coords[1] % 2 == 1:
                            self.canvas.move(self.moved[0], 1143 - min_x, 200 - min_y)
                    elif (self.give_coords[1] + 1) // 2 == 3:
                        if self.give_coords[1] % 2 == 0:
                            self.canvas.move(self.moved[0], 1291 - min_x, 200 - min_y)
                        elif self.give_coords[1] % 2 == 1:
                            self.canvas.move(self.moved[0], 1229 - min_x, 200 - min_y)
                    elif (self.give_coords[1] + 1) // 2 == 4:
                        if self.give_coords[1] % 2 == 0:
                            self.canvas.move(self.moved[0], 1415 - min_x, 200 - min_y)
                        elif self.give_coords[1] % 2 == 1:
                            self.canvas.move(self.moved[0], 1353 - min_x, 200 - min_y)
    def calculate_remaining_area(self):
        total_area = remaining_area = 130 * 85
        for shapes in self.put:
            if shapes[1] == "square":
                remaining_area -= 16
            elif shapes[1] == "hexagon":
                remaining_area -= 83
            elif shapes[1] == "square_cut":
                remaining_area -= 8
            elif shapes[1] == "hexagon_cut":
                remaining_area -= 41
        return remaining_area, int((remaining_area / total_area) * 100)
    def reset_random(self):
        self.random_button.configure(state=tk.NORMAL)
        self.stochastic_button.configure(state=tk.NORMAL)
        collision = self.canvas.find_overlapping(0, 0, 794, 1247)
        for i in range(len(collision) - 1, 0, -1):
            self.canvas.delete(collision[i])
        self.put.clear()
        self.canvas.delete(self.finished)
        ra = self.calculate_remaining_area()
        self.canvas.itemconfig(self.text_ra, text='Remaining area: ' + str(ra[0]))
        self.canvas.itemconfig(self.text_ra_per, text='Remaining area percentage: ' + str(ra[1]))
        self.canvas.update()
    def random_put(self, threshold = 100000):
        """
            Puts the tiles randomly.
        """
        trial = 0
        while trial < threshold:
            self.random_button.configure(state=tk.DISABLED) 
            self.stochastic_button.configure(state=tk.DISABLED)
            encode_shape = rd.randint(0, 1)
            encode_cut = rd.randint(0, 8)
            random_x = rd.randint(0, 793)
            random_y = rd.randint(0, 1246)
            self.moved = self.kinds[encode_shape][encode_cut]
            self.give_coords = [encode_shape, encode_cut]
            if encode_shape == 0 and encode_cut == 0:
                self.canvas.move(self.moved[0], random_x - 900 , random_y - 100)
            elif encode_shape == 0 and encode_cut != 0:
                if (self.give_coords[1] + 1) // 2 == 1:
                    if self.give_coords[1] % 2 == 0:
                        self.canvas.move(self.moved[0], random_x - 1000, random_y - 129)
                    elif self.give_coords[1] % 2 == 1:
                        self.canvas.move(self.moved[0], random_x - 1000, random_y - 100)
                elif (self.give_coords[1] + 1) // 2 == 2:
                    if self.give_coords[1] % 2 == 0:
                        self.canvas.move(self.moved[0], random_x - 1139, random_y - 100)
                    elif self.give_coords[1] % 2 == 1:
                        self.canvas.move(self.moved[0], random_x - 1100, random_y - 100)
                elif (self.give_coords[1] + 1) // 2 == 3:
                    if self.give_coords[1] % 2 == 0:
                        self.canvas.move(self.moved[0], random_x - 1239, random_y - 100)
                    elif self.give_coords[1] % 2 == 1:
                        self.canvas.move(self.moved[0], random_x - 1200, random_y - 100)
                elif (self.give_coords[1] + 1) // 2 == 4:
                    if self.give_coords[1] % 2 == 0:
                        self.canvas.move(self.moved[0], random_x - 1339, random_y - 100)
                    elif self.give_coords[1] % 2 == 1:
                        self.canvas.move(self.moved[0], random_x - 1300, random_y - 100)
            elif encode_shape == 1 and encode_cut == 0:
                self.canvas.move(self.moved[0], random_x - 881 , random_y - 200)
            elif encode_shape == 1 and encode_cut != 0:
                if (self.give_coords[1] + 1) // 2 == 1:
                    if self.give_coords[1] % 2 == 0:
                        self.canvas.move(self.moved[0], random_x - 1062, random_y - 234)
                    elif self.give_coords[1] % 2 == 1:
                        self.canvas.move(self.moved[0], random_x - 981, random_y - 200)
                elif (self.give_coords[1] + 1) // 2 == 2:
                    if self.give_coords[1] % 2 == 0:
                        self.canvas.move(self.moved[0], random_x - 1186, random_y - 200)
                    elif self.give_coords[1] % 2 == 1:
                        self.canvas.move(self.moved[0], random_x - 1143, random_y - 200)
                elif (self.give_coords[1] + 1) // 2 == 3:
                    if self.give_coords[1] % 2 == 0:
                        self.canvas.move(self.moved[0], random_x - 1291, random_y - 200)
                    elif self.give_coords[1] % 2 == 1:
                        self.canvas.move(self.moved[0], random_x - 1229, random_y - 200)
                elif (self.give_coords[1] + 1) // 2 == 4:
                    if self.give_coords[1] % 2 == 0:
                        self.canvas.move(self.moved[0], random_x - 1415, random_y - 200)
                    elif self.give_coords[1] % 2 == 1:
                        self.canvas.move(self.moved[0], random_x - 1353, random_y - 200)
            coords = self.canvas.coords(self.moved[0])
            max_x = max(coords[i] for i in range(0, len(coords), 2))
            min_x = min(coords[i] for i in range(0, len(coords), 2))
            max_y = max(coords[i] for i in range(1, len(coords), 2))
            min_y = min(coords[i] for i in range(1, len(coords), 2))
            collision = self.canvas.find_overlapping(min_x, min_y, max_x, max_y)
            if 0 <= min_x and max_x <= 794 and 0 <= min_y and max_y <= 1247 and len(collision) == 2:
                if self.moved[1] == "square" or self.moved[1] == "square_cut":
                    if self.give_coords[1] == 0:
                        shape = self.create_square()
                        self.kinds[0][0] = shape
                    else:
                        shape = self.cut_square(axis = (self.give_coords[1] + 1) // 2)[(self.give_coords[1] + 1) % 2]
                        self.kinds[0][self.give_coords[1]] = shape
                elif self.moved[1] == "hexagon" or self.moved[1] == "hexagon_cut":
                    if self.give_coords[1] == 0:
                        shape = self.create_hexagon()
                        self.kinds[1][0] = shape
                    else:
                        shape = self.cut_hexagon(axis = (self.give_coords[1] + 1) // 2)[(self.give_coords[1] + 1) % 2]
                        self.kinds[1][self.give_coords[1]] = shape
                self.put.append(self.moved)
                ra = self.calculate_remaining_area()
                self.canvas.itemconfig(self.text_ra, text='Remaining area: ' + str(ra[0]))
                self.canvas.itemconfig(self.text_ra_per, text='Remaining area percentage: ' + str(ra[1]))
                self.canvas.update()
                trial = 0
            else:
                if encode_shape == 0 and encode_cut == 0:
                    self.canvas.move(self.moved[0], 900 - min_x, 100 - min_y)
                elif encode_shape == 0 and encode_cut != 0:
                    if (self.give_coords[1] + 1) // 2 == 1:
                        if self.give_coords[1] % 2 == 0:
                            self.canvas.move(self.moved[0], 1000 - min_x, 129 - min_y)
                        elif self.give_coords[1] % 2 == 1:
                            self.canvas.move(self.moved[0], 1000 - min_x, 100 - min_y)
                    elif (self.give_coords[1] + 1) // 2 == 2:
                        if self.give_coords[1] % 2 == 0:
                            self.canvas.move(self.moved[0], 1139 - min_x, 100 - min_y)
                        elif self.give_coords[1] % 2 == 1:
                            self.canvas.move(self.moved[0], 1100 - min_x, 100 - min_y)
                    elif (self.give_coords[1] + 1) // 2 == 3:
                        if self.give_coords[1] % 2 == 0:
                            self.canvas.move(self.moved[0], 1239 - min_x, 100 - min_y)
                        elif self.give_coords[1] % 2 == 1:
                            self.canvas.move(self.moved[0], 1200 - min_x, 100 - min_y)
                    elif (self.give_coords[1] + 1) // 2 == 4:
                        if self.give_coords[1] % 2 == 0:
                            self.canvas.move(self.moved[0], 1339 - min_x, 100 - min_y)
                        elif self.give_coords[1] % 2 == 1:
                            self.canvas.move(self.moved[0], 1300 - min_x, 100 - min_y)
                elif encode_shape == 1 and encode_cut == 0:
                    self.canvas.move(self.moved[0], 881 - min_x, 200 - min_y)
                elif encode_shape == 1 and encode_cut != 0:
                    if (self.give_coords[1] + 1) // 2 == 1:
                        if self.give_coords[1] % 2 == 0:
                            self.canvas.move(self.moved[0], 1062 - min_x, 234 - min_y)
                        elif self.give_coords[1] % 2 == 1:
                            self.canvas.move(self.moved[0], 981 - min_x, 200 - min_y)
                    elif (self.give_coords[1] + 1) // 2 == 2:
                        if self.give_coords[1] % 2 == 0:
                            self.canvas.move(self.moved[0], 1186 - min_x, 200 - min_y)
                        elif self.give_coords[1] % 2 == 1:
                            self.canvas.move(self.moved[0], 1143 - min_x, 200 - min_y)
                    elif (self.give_coords[1] + 1) // 2 == 3:
                        if self.give_coords[1] % 2 == 0:
                            self.canvas.move(self.moved[0], 1291 - min_x, 200 - min_y)
                        elif self.give_coords[1] % 2 == 1:
                            self.canvas.move(self.moved[0], 1229 - min_x, 200 - min_y)
                    elif (self.give_coords[1] + 1) // 2 == 4:
                        if self.give_coords[1] % 2 == 0:
                            self.canvas.move(self.moved[0], 1415 - min_x, 200 - min_y)
                        elif self.give_coords[1] % 2 == 1:
                            self.canvas.move(self.moved[0], 1353 - min_x, 200 - min_y)
                trial += 1
        self.finished = self.canvas.create_text(900, 600, text='Random process is finished !') 
        root.after(1000, self.reset_random) 
    def stochastic_put(self, threshold = 100000):
        """
            Puts the tiles usşng the prior information.
        """
        trial = 0
        max_trial = 0
        while trial < threshold:
            self.stochastic_button.configure(state=tk.DISABLED) 
            self.random_button.configure(state=tk.DISABLED)
            encode_shape = rd.randint(0, 1)
            encode_cut_square = rd.randint(1, 8)
            random_x = rd.randint(0, 793)
            random_y = rd.randint(0, 1246)
            if max_trial < threshold // 4:
                self.moved = self.kinds[encode_shape][0]
                self.give_coords = [encode_shape, 0]
                if encode_shape == 0:
                    self.canvas.move(self.moved[0], random_x - 900 , random_y - 100)
                elif encode_shape == 1:
                    self.canvas.move(self.moved[0], random_x - 881 , random_y - 200)
                coords = self.canvas.coords(self.moved[0])
                max_x = max(coords[i] for i in range(0, len(coords), 2))
                min_x = min(coords[i] for i in range(0, len(coords), 2))
                max_y = max(coords[i] for i in range(1, len(coords), 2))
                min_y = min(coords[i] for i in range(1, len(coords), 2))
                if self.moved[1] == "hexagon":
                    collision = self.canvas.find_overlapping(min_x + 9, min_y, max_x - 9, max_y)
                else:
                    collision = self.canvas.find_overlapping(min_x, min_y, max_x, max_y)
                if (0 <= min_x and max_x <= 794 and 0 <= min_y and max_y <= 1247 and len(collision) == 2):
                    if self.moved[1] == "square":
                        shape = self.create_square()
                        self.kinds[0][0] = shape
                    elif self.moved[1] == "hexagon":
                        shape = self.create_hexagon()
                        self.kinds[1][0] = shape
                    self.put.append(self.moved)
                    ra = self.calculate_remaining_area()
                    self.canvas.itemconfig(self.text_ra, text='Remaining area: ' + str(ra[0]))
                    self.canvas.itemconfig(self.text_ra_per, text='Remaining area percentage: ' + str(ra[1]))
                    self.canvas.update()
                    trial = 0
                else:
                    if encode_shape == 0:
                        self.canvas.move(self.moved[0], 900 - min_x, 100 - min_y)
                    elif encode_shape == 1:
                        self.canvas.move(self.moved[0], 881 - min_x, 200 - min_y)
                    trial += 1
                if max_trial < trial:
                    max_trial = trial
            else:
                self.moved = self.kinds[0][encode_cut_square]
                self.give_coords = [0, encode_cut_square]
                if (self.give_coords[1] + 1) // 2 == 1:
                    if self.give_coords[1] % 2 == 0:
                        self.canvas.move(self.moved[0], random_x - 1000, random_y - 129)
                    elif self.give_coords[1] % 2 == 1:
                        self.canvas.move(self.moved[0], random_x - 1000, random_y - 100)
                elif (self.give_coords[1] + 1) // 2 == 2:
                    if self.give_coords[1] % 2 == 0:
                        self.canvas.move(self.moved[0], random_x - 1139, random_y - 100)
                    elif self.give_coords[1] % 2 == 1:
                        self.canvas.move(self.moved[0], random_x - 1100, random_y - 100)
                elif (self.give_coords[1] + 1) // 2 == 3:
                    if self.give_coords[1] % 2 == 0:
                        self.canvas.move(self.moved[0], random_x - 1239, random_y - 100)
                    elif self.give_coords[1] % 2 == 1:
                        self.canvas.move(self.moved[0], random_x - 1200, random_y - 100)
                elif (self.give_coords[1] + 1) // 2 == 4:
                    if self.give_coords[1] % 2 == 0:
                        self.canvas.move(self.moved[0], random_x - 1339, random_y - 100)
                    elif self.give_coords[1] % 2 == 1:
                        self.canvas.move(self.moved[0], random_x - 1300, random_y - 100)
                coords = self.canvas.coords(self.moved[0])
                max_x = max(coords[i] for i in range(0, len(coords), 2))
                min_x = min(coords[i] for i in range(0, len(coords), 2))
                max_y = max(coords[i] for i in range(1, len(coords), 2))
                min_y = min(coords[i] for i in range(1, len(coords), 2))
                collision = self.canvas.find_overlapping(min_x, min_y, max_x, max_y)
                if 0 <= min_x and max_x <= 794 and 0 <= min_y and max_y <= 1247 and len(collision) == 2:
                    shape = self.cut_square(axis = (self.give_coords[1] + 1) // 2)[(self.give_coords[1] + 1) % 2]
                    self.kinds[0][self.give_coords[1]] = shape
                    self.put.append(self.moved)
                    ra = self.calculate_remaining_area()
                    self.canvas.itemconfig(self.text_ra, text='Remaining area: ' + str(ra[0]))
                    self.canvas.itemconfig(self.text_ra_per, text='Remaining area percentage: ' + str(ra[1]))
                    self.canvas.update()
                    trial = 0
                else:
                    if (self.give_coords[1] + 1) // 2 == 1:
                        if self.give_coords[1] % 2 == 0:
                            self.canvas.move(self.moved[0], 1000 - min_x, 129 - min_y)
                        elif self.give_coords[1] % 2 == 1:
                            self.canvas.move(self.moved[0], 1000 - min_x, 100 - min_y)
                    elif (self.give_coords[1] + 1) // 2 == 2:
                        if self.give_coords[1] % 2 == 0:
                            self.canvas.move(self.moved[0], 1139 - min_x, 100 - min_y)
                        elif self.give_coords[1] % 2 == 1:
                            self.canvas.move(self.moved[0], 1100 - min_x, 100 - min_y)
                    elif (self.give_coords[1] + 1) // 2 == 3:
                        if self.give_coords[1] % 2 == 0:
                            self.canvas.move(self.moved[0], 1239 - min_x, 100 - min_y)
                        elif self.give_coords[1] % 2 == 1:
                            self.canvas.move(self.moved[0], 1200 - min_x, 100 - min_y)
                    elif (self.give_coords[1] + 1) // 2 == 4:
                        if self.give_coords[1] % 2 == 0:
                            self.canvas.move(self.moved[0], 1339 - min_x, 100 - min_y)
                        elif self.give_coords[1] % 2 == 1:
                            self.canvas.move(self.moved[0], 1300 - min_x, 100 - min_y)
                    trial += 1
        self.finished = self.canvas.create_text(900, 600, text='Stochastic process is finished !') 
        root.after(1000, self.reset_random)
if __name__== '__main__':
    root=tk.Tk()
    gui = TilesGui(root)
    root.mainloop()
 
```
